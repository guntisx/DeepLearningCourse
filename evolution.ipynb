```python
# evolutionary learning example: minimize loss around some target point
import numpy as np
target = np.array([0.5, 0.1, -0.3])
def f(w): return np.sum((w - target)**2)
loss_values = []

npop = 50      # population size
sigma = 0.1    # mutation standard deviation
alpha = 0.001  # learning rate
w = np.random.randn(3) # initial guess
for i in range(300):
  N = np.random.randn(npop, 3)
  R = np.zeros(npop)
  for j in range(npop):
    w_try = w + sigma*N[j]
    R[j] = f(w_try)
  A = (np.mean(R) - R) / np.std(R)
  w = w + alpha/(npop*sigma) * np.dot(N.T, A)
  #print (i, np.mean(R), f(w))
  loss_values.append(f(w))
print ("target: ",target)
print ("solution: ",w)

import matplotlib.pyplot as plt
plt.plot(loss_values)
plt.show()
```
